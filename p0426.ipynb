```python
# seaborn, SNS 특집
import seaborn as sns

anscombe = sns.load_dataset('anscombe')

print(anscombe)
print(type(anscombe))
```

```python
import matplotlib.pyplot as plt

dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_1
```

```python
# 기본 그리기

plt.plot(dataset_1['x'], dataset_1['y'], 'o')
# plot은 기본적으로 선으로 그래프를 그린다. 세번째 인수로
```

```python
dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
# 기본틀, 전체틀(액자) 생성

fig = plt.figure()

axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')
```

```python
fig = plt.figure()
fig.suptitle('Anscombe')


axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

fig.tight_layout() # 그래프 이름과 숫자가 겹쳐질 때 크기를 조절
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
print(str(tips))

# tips 지불금액, 팁, 성별, 흡연여부, 요일, 시간, 인원수
```

```python
# fig 기본틀 액자
fig = plt.figure()
# 기본틀에 내부틀 -> 1개짜리
axes1 = fig.add_subplot(1,1,1)

axes1.hist(tips['total_bill'])

axes1.set_title('Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Bill')
# total bill 최종 계산 가격
plt.show()
```

```python
scatter_plot = plt.figure()

axes1 = scatter_plot.add_subplot(1,1,1)

axes1.scatter(tips['total_bill'], tips['tip'])

axes1.set_title('Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

plt.show()
```

```python
# 박스 그래프 - 성별과 팁의 상관관계

boxplot = plt.figure()

axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == "Female"]['tip'],
               tips[tips['sex'] == "Male"]['tip']],
               labels=['Female','Male'])

axes1.set_title('Tips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')

plt.show()
```

```python
# Seaborn 그리기
# seaborn으로 그림을 그리면 matplotlib 보다 좀 더 화려하게 그릴 수 있다.

import seaborn as sns

tips = sns.load_dataset('tips')
```

```python
ax = plt.subplot()

ax = sns.distplot(tips['total_bill'])

ax.set_title('Total Bill Histogram with SNS')

plt.show()
```

```python
ax = plt.subplot()

ax = sns.distplot(tips['total_bill'], rug=True)

ax.set_title('Total Bill Histogram with SNS')

plt.show()
```

```python
ax = plt.subplot()

ax = sns.countplot('day', data=tips)

ax.set_title('Count of Days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Count')

plt.show()
```

```python
ax = plt.subplot()

ax = sns.regplot(x='total_bill', y='tip', data=tips)

ax.set_title('Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

plt.show()
```

```python
#ax = plt.subplot()

ax = sns.jointplot(x='total_bill', y='tip', data=tips)
ax.set_axis_labels(xlabel="Total Bill", ylabel='Tip')
ax.fig.suptitle('Joint plot of Total Bill and Tip', y=1.1)

#ax.set_title('Total Bill and Tip')
#ax.set_xlabel('Total Bill')
#ax.set_ylabel('Tip')

plt.show()
```

```python
#ax = plt.subplot()

ax = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
ax.set_axis_labels(xlabel="Total Bill", ylabel='Tip')
ax.fig.suptitle('Joint plot of Total Bill and Tip', y=1.1)

#ax.set_title('Total Bill and Tip')
#ax.set_xlabel('Total Bill')
#ax.set_ylabel('Tip')

plt.show()
```

```python
ax = plt.subplot()

ax = sns.kdeplot(data=tips, data2=tips['tip'])

ax.set_title('Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

plt.show()
```

```python
ax = plt.subplot()

ax = sns.barplot(x='time', y='tip', data=tips)

ax.set_title('Time and Tip')
ax.set_xlabel('Time')
ax.set_ylabel('Tip')

plt.show()
```

```python
ax = plt.subplot()

ax = sns.boxplot(x='time', y='tip', data=tips)

ax.set_title('Time and Tip')
ax.set_xlabel('Time')
ax.set_ylabel('Tip')

plt.show()
```

```python
ax = plt.subplot()

ax = sns.violinplot(x='time', y='tip', data=tips)

ax.set_title('Time and Tip')
ax.set_xlabel('Time')
ax.set_ylabel('Tip')

plt.show()
```
